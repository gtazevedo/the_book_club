## 1. The Markov Property

### Definition

A stochastic process $\{X_t,\, t \ge 0\}$ is a **Markov process** if, for any times $t_0 < t_1 < \cdots < t_n < t_{n+1}$:

$$\boxed{P\bigl(X_{t_{n+1}} \le x \mid X_{t_0}, X_{t_1}, \ldots, X_{t_n}\bigr) = P\bigl(X_{t_{n+1}} \le x \mid X_{t_n}\bigr)}$$

In plain English: **the entire future probability distribution of $X$ is fully determined by its present value, not by any part of its history.**

---

### Why it matters for finance

The Markov property is the mathematical formalisation of the **weak form of the Efficient Market Hypothesis (EMH)**.

- **Weak EMH** says: all *past price information* is already reflected in the current price. Knowing that a stock went up three days in a row does not give you any additional predictive power beyond the price today.
- This is exactly the Markov property: the conditional distribution of the future price depends only on the current price $S_t$.
- **Implication for modelling**: we only need $S_t$ to compute option prices and hedge ratios. We do not need to track the entire price history.

---

### The memory-less property and order of a Markov process

The definition above is for a *first-order* Markov process. More generally:

- **First-order Markov**: Future $\perp$ past $\mid$ present $X_t$.
- **Second-order Markov**: Future $\perp$ past $\mid$ $(X_t, X_{t-1})$. (ARMA(1,1) models are approximately of this type.)

All standard option-pricing models (BSM, Heston, etc.) assume a *first-order* Markov process. Fractional Brownian Motion (Section 7) is a notable *non-Markovian* process.

---

### Example: Verifying the Markov property for Brownian Motion

Let $W_t$ be a standard Wiener process. We want to show it is Markov.

For any $s < t$, the increment $W_t - W_s$ is:
1. **Independent of $\mathcal{F}_s$** (the sigma-algebra of all information up to time $s$), by the independent increments property.
2. Normally distributed: $W_t - W_s \sim \mathcal{N}(0, t - s)$.

Therefore:
$$W_t = W_s + (W_t - W_s)$$

The conditional distribution of $W_t$ given $\mathcal{F}_s$ equals the conditional distribution given $W_s$ alone:
$$W_t \mid \mathcal{F}_s \sim \mathcal{N}(W_s,\; t - s) \quad \text{(depends only on } W_s\text{)}$$

This confirms $W_t$ is Markov.

---

### Variance grows linearly with time

A key consequence of the Markov (and specifically Wiener) structure is that uncertainty accumulates at a specific rate. For any Wiener process:

$$\text{Var}(W_t - W_0) = t$$

More generally, if the variance *rate* is $\sigma^2$ per unit time:

$$\text{Var}(X_t - X_0) = \sigma^2 t \implies \text{Std}(X_t - X_0) = \sigma\sqrt{t}$$

This is the origin of the famous $\sqrt{t}$ ("square-root-of-time") scaling rule used throughout risk management and option pricing.
Doubling your holding period does not double your risk — it scales it by $\sqrt{2} \approx 1.41$.

## 2. Constructing the Wiener Process as the Limit of a Random Walk

The Wiener process is not defined abstractly — it can be built concretely from a sequence of fair coin flips.

---

### Step 1 — Set up the discrete symmetric random walk

Partition the time horizon $[0, T]$ into $N$ equal steps of length $\Delta t = T/N$. At each step $i$, the process moves by a random amount:

$$X_i = \begin{cases} +\Delta x & \text{with probability } \tfrac{1}{2} \\ -\Delta x & \text{with probability } \tfrac{1}{2} \end{cases}$$

The steps $X_1, X_2, \ldots, X_N$ are independent and identically distributed (i.i.d.).

**Moments of a single step:**
$$\mathbb{E}[X_i] = \tfrac{1}{2}(+\Delta x) + \tfrac{1}{2}(-\Delta x) = 0$$
$$\mathbb{E}[X_i^2] = \tfrac{1}{2}(\Delta x)^2 + \tfrac{1}{2}(\Delta x)^2 = (\Delta x)^2$$
$$\text{Var}(X_i) = \mathbb{E}[X_i^2] - (\mathbb{E}[X_i])^2 = (\Delta x)^2$$

---

### Step 2 — Accumulate steps to form the path

The accumulated position after $k$ steps is:
$$W^{(N)}_{k\Delta t} = \sum_{i=1}^{k} X_i$$

**Mean of the total walk after $N$ steps (time $T$):**
$$\mathbb{E}\left[W^{(N)}_T\right] = \sum_{i=1}^{N} \mathbb{E}[X_i] = N \cdot 0 = 0$$

**Variance of the total walk** (independence of steps $\Rightarrow$ variances add):
$$\text{Var}\left(W^{(N)}_T\right) = \sum_{i=1}^{N} \text{Var}(X_i) = N \cdot (\Delta x)^2 = \frac{T}{\Delta t} \cdot (\Delta x)^2$$

---

### Step 3 — Choose the scaling $\Delta x = \sqrt{\Delta t}$

As $\Delta t \to 0$ (equivalently $N \to \infty$), we want the variance to remain finite and meaningful. Setting $\text{Var} = T$ requires:

$$\frac{T}{\Delta t} \cdot (\Delta x)^2 = T \implies (\Delta x)^2 = \Delta t \implies \Delta x = \pm\sqrt{\Delta t}$$

> **Insight:** Each step is of size $\sqrt{\Delta t}$, which is much larger than $\Delta t$ itself when $\Delta t$ is small. This asymmetry between the size of the time step and the size of the price step is the mathematical source of the roughness of Brownian paths. It also explains why we cannot differentiate $W_t$ in the classical sense.

---

### Step 4 — Apply the Central Limit Theorem

With $\Delta x = \pm\sqrt{\Delta t}$, the sum $W^{(N)}_T = \sum_{i=1}^{N} X_i$ consists of $N \to \infty$ i.i.d. terms, each with mean 0 and variance $\Delta t$. By the **CLT**:

$$W^{(N)}_T \xrightarrow{\;d\;} \mathcal{N}\bigl(0,\; N \cdot \Delta t\bigr) = \mathcal{N}(0,\, T)$$

More generally, for any $s < t$, the increment $W^{(N)}_t - W^{(N)}_s$ converges to $\mathcal{N}(0, t-s)$.

---

### Step 5 — Properties of the limiting Wiener process

The limit $W_t = \lim_{N \to \infty} W^{(N)}_t$ is a **standard Wiener process** (or standard Brownian Motion), characterised by:

| Property | Mathematical statement |
|----------|------------------------|
| Starts at 0 | $W_0 = 0$ |
| Continuous paths | $t \mapsto W_t$ is continuous a.s. |
| Normal increments | $W_t - W_s \sim \mathcal{N}(0,\, t-s)$ for $s < t$ |
| Independent increments | $W_t - W_s \perp \mathcal{F}_s$ for $s < t$ |
| Markov property | follows from independent increments |

---

### Step 6 — The continuous-time differential notation

In the limit, the infinitesimal increment over $[t, t+dt]$ is written:

$$dW_t = \varepsilon\,\sqrt{dt}, \quad \varepsilon \sim \mathcal{N}(0, 1) \text{ independent}$$

This gives:
$$\mathbb{E}[dW_t] = 0, \qquad \mathbb{E}[(dW_t)^2] = dt, \qquad \mathbb{E}[dW_s \cdot dW_t] = 0 \text{ for } s \ne t$$

Integrating over $[0, T]$:
$$W_T = \int_0^T dW_t \sim \mathcal{N}(0,\, T)$$

---

### Step 7 — Non-differentiability of Brownian paths

Although Brownian paths are continuous, they are **nowhere differentiable** (with probability 1). To see why, consider the ratio:

$$\frac{W_{t+h} - W_t}{h} \sim \frac{\mathcal{N}(0,\,h)}{h} = \frac{1}{\sqrt{h}} \cdot \mathcal{N}(0,1)$$

As $h \to 0$, this ratio diverges (it blows up like $h^{-1/2}$). Therefore the derivative $\frac{dW_t}{dt}$ does not exist in the classical sense. This is why we use the notation $dW$ (an increment) rather than $W'(t)$ (a derivative).

## 3. Generalized Wiener Process

The standard Wiener process $dW$ has zero drift and variance rate 1. To model variables with a **non-zero trend** and **arbitrary noise scaling**, we introduce the generalised form:

$$\boxed{dx = a\,dt + b\,dW}$$

where:
- $a$ = **drift rate** (deterministic trend; units: [variable per unit time])
- $b$ = **diffusion coefficient** (noise scale; units: [variable per $\sqrt{\text{time}}$])
- $dW$ = standard Wiener increment $\varepsilon\sqrt{dt}$

---

### Step 1 — Integrate the SDE

Integrating both sides from $0$ to $T$:
$$x(T) - x(0) = \int_0^T a\,dt + \int_0^T b\,dW = aT + b W_T$$

The first integral is a classical (Riemann) integral of a constant. The second is an **Itô integral** — but since $b$ is constant, it reduces to $b \cdot W_T$.

---

### Step 2 — Expected value

$$\mathbb{E}[x(T)] = x(0) + aT + b\,\mathbb{E}[W_T] = x(0) + aT$$

The drift rate $a$ contributes a **purely deterministic, linear trend**. If $a > 0$, the variable grows on average at rate $a$ per unit time.

---

### Step 3 — Variance

Since $aT$ is a deterministic constant (variance zero):
$$\text{Var}(x(T) - x(0)) = \text{Var}(b\,W_T) = b^2 \,\text{Var}(W_T) = b^2 T$$

The standard deviation grows as $b\sqrt{T}$. The coefficient $b$ is the **volatility** of the variable (analogous to $\sigma$ for stock prices).

---

### Step 4 — Distribution

Since $x(T) - x(0) = aT + b W_T$ is an affine (linear) transformation of the Gaussian $W_T$:

$$\boxed{x(T) - x(0) \sim \mathcal{N}\bigl(aT,\; b^2 T\bigr)}$$

Equivalently, using discrete-time notation for a short step $\Delta t$:
$$\Delta x = a\,\Delta t + b\,\varepsilon\sqrt{\Delta t}, \quad \varepsilon \sim \mathcal{N}(0,1)$$

---

### Step 5 — Hull's cash position example (Hull Example 14.2)

Hull considers an investor whose dollar holdings $x$ follow:
$$dx = 20\,dt + 30\,dW$$

Starting from $x_0 = 50$ today, what is the distribution of $x$ in 3 months ($T = 0.25$ years)?

$$\mathbb{E}[x(T)] = 50 + 20 \times 0.25 = \$55$$
$$\text{Std}(x(T)) = 30\sqrt{0.25} = 30 \times 0.5 = \$15$$
$$x(T) \sim \mathcal{N}(55,\; 15^2)$$

Interpretation: we expect to end up at \$55, but with a standard deviation of \$15. A 95% confidence interval is approximately $55 \pm 30$, i.e., $[\$25, \$85]$.

---

### Step 6 — Why a Generalized Wiener Process is too simple for stocks

The generalised Wiener process has **constant** $a$ and $b$. This creates a problem for stock prices:

1. **Absolute drift**: $dx = a\,dt$ means the expected change per year is always \$20, regardless of whether the stock is at \$10 or \$10,000. This is economically unreasonable — investors think in **percentage** returns, not absolute dollar amounts.

2. **Absolute volatility**: Similarly, $b = 30$ gives a fixed \$30 standard deviation regardless of price level. But a \$30 move on a \$50 stock (60%) is very different from a \$30 move on a \$1,000 stock (3%).

3. **Negative prices possible**: Since $x(T) \sim \mathcal{N}(\cdot)$, there is always a positive probability that $x(T) < 0$. Stock prices cannot be negative.

This motivates the Itô process, where $a$ and $b$ can depend on the current state.

## 4. The Itô Process — State-Dependent Drift and Diffusion

An **Itô process** generalises the generalised Wiener process by allowing both coefficients to depend on the current value of the process and on time:

$$\boxed{dx = a(x, t)\,dt + b(x, t)\,dW}$$

This is the most general class of stochastic differential equation (SDE) used in finance. Black-Scholes, Vasicek, CIR, Heston — every standard model is a special case.

---

### Why state-dependence is necessary for stock prices

For a stock price $S$, the natural model is:

$$dS = \mu S\,dt + \sigma S\,dW$$

Here:
- $a(S, t) = \mu S$ — the expected dollar change per unit time is proportional to the price level. This gives a constant *percentage* return $\mu$ (= expected return per year).
- $b(S, t) = \sigma S$ — the noise scales with price, giving a constant *percentage* volatility $\sigma$.

These two features — proportional drift and proportional diffusion — are what make the logarithmic return $\ln(S_T/S_0)$ normally distributed (as we will show in Section 6).

---

### Taxonomy of Itô processes used in finance

| Model | SDE | $a(x,t)$ | $b(x,t)$ |
|-------|-----|----------|----------|
| **GBM** (Black-Scholes stock) | $dS = \mu S\,dt + \sigma S\,dW$ | $\mu S$ | $\sigma S$ |
| **Vasicek** (short rate) | $dr = \kappa(\theta - r)\,dt + \sigma\,dW$ | $\kappa(\theta - r)$ | $\sigma$ |
| **CIR** (short rate) | $dr = \kappa(\theta - r)\,dt + \sigma\sqrt{r}\,dW$ | $\kappa(\theta - r)$ | $\sigma\sqrt{r}$ |
| **Heston** (stochastic vol) | $dv = \kappa(\theta - v)\,dt + \xi\sqrt{v}\,dW$ | $\kappa(\theta - v)$ | $\xi\sqrt{v}$ |
| **Ornstein-Uhlenbeck** | $dx = \kappa(\theta - x)\,dt + \sigma\,dW$ | $\kappa(\theta - x)$ | $\sigma$ |

---

### Existence and uniqueness of solutions (Lipschitz conditions)

For the SDE $dx = a(x,t)\,dt + b(x,t)\,dW$ to have a **unique strong solution**, it is sufficient (but not necessary) that $a$ and $b$ satisfy:

1. **Lipschitz condition** (controls how fast coefficients vary):
$$|a(x,t) - a(y,t)| + |b(x,t) - b(y,t)| \le K|x - y|$$

2. **Linear growth condition** (prevents blow-up in finite time):
$$|a(x,t)|^2 + |b(x,t)|^2 \le K^2(1 + |x|^2)$$

GBM and most finance models satisfy these conditions away from $x = 0$.

---

### The Ornstein-Uhlenbeck process — a worked example

The OU process $dx = \kappa(\theta - x)\,dt + \sigma\,dW$ can be solved exactly.

**Step 1 — Rewrite as a linear ODE perturbed by noise:**
$$dx + \kappa x\,dt = \kappa\theta\,dt + \sigma\,dW$$

**Step 2 — Multiply by the integrating factor $e^{\kappa t}$:**
$$d(e^{\kappa t} x) = \kappa\theta\,e^{\kappa t}\,dt + \sigma\,e^{\kappa t}\,dW$$

**Step 3 — Integrate from 0 to $t$:**
$$e^{\kappa t}x_t - x_0 = \kappa\theta\int_0^t e^{\kappa s}\,ds + \sigma\int_0^t e^{\kappa s}\,dW_s$$
$$= \theta(e^{\kappa t} - 1) + \sigma\int_0^t e^{\kappa s}\,dW_s$$

**Step 4 — Solve for $x_t$:**
$$x_t = \theta + (x_0 - \theta)e^{-\kappa t} + \sigma e^{-\kappa t}\int_0^t e^{\kappa s}\,dW_s$$

**Interpretation:**
- The mean reverts to $\theta$: $\mathbb{E}[x_t] = \theta + (x_0 - \theta)e^{-\kappa t}$. As $t \to \infty$, $\mathbb{E}[x_t] \to \theta$.
- The variance: $\text{Var}(x_t) = \frac{\sigma^2}{2\kappa}(1 - e^{-2\kappa t}) \to \frac{\sigma^2}{2\kappa}$ as $t \to \infty$.
- The long-run distribution is $\mathcal{N}\!\left(\theta,\, \frac{\sigma^2}{2\kappa}\right)$.

## 5. Derivation of Itô's Lemma via Taylor Expansion

Itô's Lemma is the chain rule of stochastic calculus. In ordinary calculus, if $G$ is a smooth function and $x$ changes by $dx$, then $dG = G'(x)\,dx$. In stochastic calculus, this is **wrong** — an extra term appears because $(dW)^2 = dt \ne 0$.

**Setup.** Let $x$ follow the Itô process:
$$dx = a(x,t)\,dt + b(x,t)\,dW$$

Let $G(x, t)$ be twice continuously differentiable in $x$ and once in $t$. We seek $dG$.

---

### Step 1 — Write the full 2D Taylor expansion

$$G(x + dx,\; t + dt) = G + \frac{\partial G}{\partial x}dx + \frac{\partial G}{\partial t}dt
+ \frac{1}{2}\frac{\partial^2 G}{\partial x^2}(dx)^2 + \frac{\partial^2 G}{\partial x\,\partial t}\,dx\,dt
+ \frac{1}{2}\frac{\partial^2 G}{\partial t^2}(dt)^2 + \cdots$$

So:
$$dG = \frac{\partial G}{\partial x}dx + \frac{\partial G}{\partial t}dt
+ \frac{1}{2}\frac{\partial^2 G}{\partial x^2}(dx)^2
+ \underbrace{\frac{\partial^2 G}{\partial x\,\partial t}\,dx\,dt
+ \frac{1}{2}\frac{\partial^2 G}{\partial t^2}(dt)^2}_{\text{higher-order: go to 0 faster than }dt} + \cdots$$

---

### Step 2 — Expand $(dx)^2$

Substitute $dx = a\,dt + b\,dW$:
$$(dx)^2 = (a\,dt + b\,dW)^2 = a^2(dt)^2 + 2ab\,(dt)(dW) + b^2(dW)^2$$

---

### Step 3 — The stochastic multiplication table

This is the **key step** unique to stochastic calculus. We evaluate the order of each second-order term:

| Term | Expression | Order | Limit as $dt \to 0$ |
|------|-----------|-------|---------------------|
| $(dt)^2$ | $(dt)^2$ | $O(dt^2)$ | $\to 0$ |
| $dt \cdot dW$ | $dt \cdot \varepsilon\sqrt{dt} = \varepsilon\,dt^{3/2}$ | $O(dt^{3/2})$ | $\to 0$ |
| $(dW)^2$ | $\varepsilon^2 \cdot dt$ | $O(dt)$ | **$\to dt$ (non-zero!)** |

**Why does $(dW)^2 \to dt$?**

Since $dW = \varepsilon\sqrt{dt}$ with $\varepsilon \sim \mathcal{N}(0,1)$:
$$(dW)^2 = \varepsilon^2 \cdot dt$$

The random variable $\varepsilon^2$ has:
$$\mathbb{E}[\varepsilon^2] = 1 \quad (\text{since } \varepsilon \sim \mathcal{N}(0,1))$$
$$\text{Var}(\varepsilon^2) = \mathbb{E}[\varepsilon^4] - (\mathbb{E}[\varepsilon^2])^2 = 3 - 1 = 2$$

Therefore:
$$\mathbb{E}[(dW)^2] = \mathbb{E}[\varepsilon^2] \cdot dt = dt$$
$$\text{Var}((dW)^2) = \text{Var}(\varepsilon^2) \cdot (dt)^2 = 2(dt)^2 \xrightarrow{dt \to 0} 0$$

The variance of $(dW)^2$ vanishes faster than its mean. By the law of large numbers applied to the sum of many such terms, $(dW)^2$ **concentrates** on its mean $dt$.

This is formalised by the **quadratic variation** of Brownian motion:
$$[W, W]_t = \lim_{\|\Pi\| \to 0} \sum_{i} (W_{t_{i+1}} - W_{t_i})^2 = t \quad \text{(a.s.)}$$

Compare: for a differentiable function $f$, the analogous sum would be:
$$\sum_i (f(t_{i+1}) - f(t_i))^2 \approx \sum_i (f'(t_i))^2 (\Delta t)^2 \to 0$$
The Brownian quadratic variation is non-zero precisely because Brownian paths are not differentiable.

**The stochastic Itô multiplication table:**
$$dW \cdot dW = dt, \qquad dt \cdot dW = 0, \qquad dt \cdot dt = 0$$

---

### Step 4 — Substitute back into $(dx)^2$

$$(dx)^2 = a^2\underbrace{(dt)^2}_{=\,0} + 2ab\underbrace{\,dt\,dW}_{=\,0} + b^2\underbrace{(dW)^2}_{=\,dt} = b^2\,dt$$

---

### Step 5 — Substitute $dx$ and $(dx)^2$ into the Taylor expansion

Drop all $O(dt^{3/2})$ and higher terms:
$$dG = \frac{\partial G}{\partial x}(a\,dt + b\,dW) + \frac{\partial G}{\partial t}\,dt + \frac{1}{2}\frac{\partial^2 G}{\partial x^2}\,b^2\,dt$$

---

### Step 6 — Collect $dt$ and $dW$ terms

$$\boxed{dG = \left(\frac{\partial G}{\partial x}\,a + \frac{\partial G}{\partial t} + \frac{1}{2}\frac{\partial^2 G}{\partial x^2}\,b^2\right)dt + \frac{\partial G}{\partial x}\,b\,dW}$$

This is **Itô's Lemma**.

> **The extra term**: $\frac{1}{2}\frac{\partial^2 G}{\partial x^2}b^2$ is entirely absent in ordinary calculus. It arises because the second-order term $(dx)^2 = b^2\,dt$ in the Taylor expansion is *not* negligible when $x$ follows a stochastic process. This term is the mathematical signature of stochasticity.

---

### Sanity check: Itô's Lemma reduces to the chain rule when $b = 0$

If $b = 0$, then $x$ is deterministic ($dx = a\,dt$) and Itô's Lemma becomes:
$$dG = \left(\frac{\partial G}{\partial x}\,a + \frac{\partial G}{\partial t}\right)dt$$

This is exactly the classical chain rule (total derivative) from ordinary calculus. $\checkmark$

---

### Itô's Lemma in vector form (multi-dimensional case)

When $x = (x_1, \ldots, x_n)$ follows a system of Itô processes:
$$dx_i = a_i(\mathbf{x}, t)\,dt + \sum_{j=1}^m b_{ij}(\mathbf{x}, t)\,dW^{(j)}_t$$

Then for $G(\mathbf{x}, t)$:
$$dG = \frac{\partial G}{\partial t}\,dt + \sum_i \frac{\partial G}{\partial x_i}\,dx_i + \frac{1}{2}\sum_{i,k}\frac{\partial^2 G}{\partial x_i\,\partial x_k}\,d[x_i, x_k]_t$$

where $d[x_i, x_k]_t = \sum_j b_{ij}b_{kj}\,dt$ is the quadratic covariation.

The Heston model uses this 2D version with correlated Brownian motions $(W^S, W^v)$.

## 6. Exact Solution of Geometric Brownian Motion

The standard model for a non-dividend-paying stock price is:
$$\boxed{dS = \mu S\,dt + \sigma S\,dW} \quad \text{(Geometric Brownian Motion)}$$

Here:
- $\mu$ = continuously compounded expected return (per year)
- $\sigma$ = volatility (standard deviation of return per year)
- $S$ = stock price (always positive)

We solve this SDE explicitly using Itô's Lemma.

---

### Step 1 — Choose the transformation $G(S, t) = \ln S$

In ordinary calculus, $d(\ln S) = dS/S$. In stochastic calculus this must be corrected. We apply Itô's Lemma with $x = S$, $a = \mu S$, $b = \sigma S$:

Compute the partial derivatives:
$$\frac{\partial G}{\partial S} = \frac{1}{S}, \quad \frac{\partial^2 G}{\partial S^2} = -\frac{1}{S^2}, \quad \frac{\partial G}{\partial t} = 0$$

---

### Step 2 — Plug into Itô's Lemma

$$d(\ln S) = \left(\frac{1}{S}\cdot\mu S + 0 + \frac{1}{2}\cdot\left(-\frac{1}{S^2}\right)\cdot\sigma^2 S^2\right)dt + \frac{1}{S}\cdot\sigma S\,dW$$

Simplify each term:
$$d(\ln S) = \left(\mu - \frac{\sigma^2}{2}\right)dt + \sigma\,dW$$

**Crucial observation:** The right-hand side has constant coefficients — it is a Generalized Wiener Process! The Itô correction $-\sigma^2/2$ has absorbed all the state-dependence.

---

### Step 3 — Why the Itô correction $-\sigma^2/2$ appears

Without stochastic calculus (i.e., if we naively applied the classical chain rule):
$$d(\ln S) \stackrel{?}{=} \frac{dS}{S} = \mu\,dt + \sigma\,dW$$

But with Itô's Lemma, the extra term $\frac{1}{2}\frac{\partial^2(\ln S)}{\partial S^2}(\sigma S)^2 = \frac{1}{2}\left(-\frac{1}{S^2}\right)\sigma^2 S^2 = -\frac{\sigma^2}{2}$ is subtracted.

This correction is the consequence of the **convexity of $\ln S$**: because $\ln$ is concave, Jensen's inequality implies $\mathbb{E}[\ln S_T] < \ln \mathbb{E}[S_T]$, and the difference is precisely $\frac{\sigma^2}{2}T$.

---

### Step 4 — Integrate both sides from $0$ to $T$

$$\int_0^T d(\ln S_u) = \int_0^T \left(\mu - \frac{\sigma^2}{2}\right)du + \int_0^T \sigma\,dW_u$$
$$\ln S_T - \ln S_0 = \left(\mu - \frac{\sigma^2}{2}\right)T + \sigma W_T$$

---

### Step 5 — Exponentiate to recover $S_T$

$$\boxed{S_T = S_0\,\exp\!\left[\left(\mu - \frac{\sigma^2}{2}\right)T + \sigma W_T\right]}$$

Since $W_T \sim \mathcal{N}(0, T)$, the log-return is:
$$\ln\frac{S_T}{S_0} \sim \mathcal{N}\!\left(\left(\mu - \frac{\sigma^2}{2}\right)T,\; \sigma^2 T\right)$$

and $S_T$ itself follows a **lognormal distribution**.

---

### Step 6 — The variance drag $-\sigma^2/2$ — a deeper look

**Why is the mean of $\ln(S_T/S_0)$ equal to $(\mu - \sigma^2/2)T$, not $\mu T$?**

Consider a simple two-period example: Starting at \$100, the stock goes up 20% then down 20%.
- Up: $100 \times 1.20 = \$120$
- Down: $120 \times 0.80 = \$96$

The arithmetic average return is $(+20\% + -20\%)/2 = 0\%$, but the realised final value is \$96, not \$100. This difference is the variance drag:
$$\text{Geometric mean} \approx \text{Arithmetic mean} - \frac{\sigma^2}{2}$$

More formally, by Jensen's Inequality (since $\ln$ is concave):
$$\mathbb{E}[\ln S_T] \le \ln\mathbb{E}[S_T]$$

with:
$$\mathbb{E}[S_T] = S_0 e^{\mu T} \quad \text{but} \quad \mathbb{E}[\ln S_T] = \ln S_0 + \left(\mu - \frac{\sigma^2}{2}\right)T$$

The gap $\sigma^2 T / 2$ is the **variance drag** (also called the volatility drag). Higher volatility $\sigma$ reduces the geometric mean return, even if the arithmetic expected return $\mu$ stays constant. This has profound implications for long-horizon investors.

---

### Step 7 — The lognormal distribution and its moments

If $\ln(S_T/S_0) \sim \mathcal{N}(m, v)$ with $m = (\mu - \sigma^2/2)T$ and $v = \sigma^2 T$, then:

$$\mathbb{E}[S_T] = S_0\,e^{m + v/2} = S_0\,e^{(\mu - \sigma^2/2)T + \sigma^2 T/2} = S_0 e^{\mu T}$$

$$\text{Var}(S_T) = S_0^2 e^{2m + v}(e^v - 1) = S_0^2 e^{2\mu T}(e^{\sigma^2 T} - 1)$$

The expected stock price grows at the *arithmetic* rate $\mu$, confirming consistency with the SDE $\mathbb{E}[dS] = \mu S\,dt$. Meanwhile, the *geometric* (compounded) growth rate is $\mu - \sigma^2/2$.

---

### Step 8 — Simulating GBM: Euler-Maruyama vs Exact

Two common discretisation schemes for GBM:

**Euler-Maruyama** (discretise the SDE directly):
$$S_{t+\Delta t} \approx S_t + \mu S_t \Delta t + \sigma S_t \varepsilon\sqrt{\Delta t}$$
This introduces discretisation error and can occasionally produce negative prices.

**Exact simulation** (use the closed-form solution):
$$S_{t+\Delta t} = S_t \exp\!\left[\left(\mu - \frac{\sigma^2}{2}\right)\Delta t + \sigma\varepsilon\sqrt{\Delta t}\right]$$
This is **exact** (no discretisation error) and always produces positive prices. It should always be preferred for GBM.


---
## 7. Fractional Brownian Motion: Covariance Structure and Hurst Index

Standard Brownian Motion assumes independent increments: knowing yesterday's return tells you nothing about tomorrow's. This is the Markov property. **Fractional Brownian Motion (fBm)** relaxes this by allowing increments to be positively or negatively correlated — capturing **long memory** (trending markets, $H > 1/2$) or **short memory / mean-reversion** (choppy markets, $H < 1/2$).

---

### Definition

Fractional Brownian Motion $B_H(t)$ is the unique (up to scaling) Gaussian process such that:

1. $B_H(0) = 0$
2. $\mathbb{E}[B_H(t)] = 0$ for all $t$ (zero mean)
3. **Stationary increments**: the distribution of $B_H(t+s) - B_H(t)$ depends only on $s$
4. **Self-similarity** with Hurst exponent $H \in (0,1)$: $\{B_H(ct)\} \overset{d}{=} \{c^H B_H(t)\}$ for all $c > 0$

When $H = 1/2$, $B_{1/2}$ is standard Brownian Motion.

---

### Step 1 — Variance of increments from self-similarity

Setting $c = t$:
$$B_H(t) \overset{d}{=} t^H B_H(1) \implies \text{Var}(B_H(t)) = t^{2H}$$

(normalising so $\text{Var}(B_H(1)) = 1$)

By stationarity of increments:
$$\text{Var}(B_H(t) - B_H(s)) = |t - s|^{2H}$$

For $H = 1/2$: $\text{Var} = |t-s|$, consistent with standard BM. $\checkmark$

---

### Step 2 — Derive the covariance function

For $t, s \ge 0$ (assume WLOG $t \ge s$):

**Expand the variance of the increment:**
$$|t - s|^{2H} = \text{Var}(B_H(t) - B_H(s)) = t^{2H} + s^{2H} - 2\mathbb{E}[B_H(t)B_H(s)]$$

(using $\mathbb{E}[(A-B)^2] = \mathbb{E}[A^2] - 2\mathbb{E}[AB] + \mathbb{E}[B^2]$ and zero means)

**Rearrange:**
$$\boxed{\mathbb{E}[B_H(t)B_H(s)] = \frac{1}{2}\left(t^{2H} + s^{2H} - |t-s|^{2H}\right)}$$

---

### Step 3 — Covariance of consecutive unit increments

Let $\Delta_k = B_H(k) - B_H(k-1)$ be the $k$-th unit increment. We compute $\text{Cov}(\Delta_1, \Delta_2)$:

$$\text{Cov}(\Delta_1, \Delta_2) = \mathbb{E}[\Delta_1 \Delta_2] = \mathbb{E}[B_H(1)B_H(2)] - \mathbb{E}[B_H(1)^2] - 0 + 0$$

Using the covariance formula:
$$\mathbb{E}[B_H(1)B_H(2)] = \frac{1}{2}(1 + 2^{2H} - 1) = \frac{2^{2H}}{2} = 2^{2H-1}$$
$$\mathbb{E}[B_H(1)^2] = 1$$

$$\text{Cov}(\Delta_1, \Delta_2) = 2^{2H-1} - 1 = \frac{4^H - 2}{2}$$

---

### Step 4 — The role of the Hurst index

| $H$ | Covariance sign | Behaviour | Real market analogy |
|-----|-----------------|-----------|---------------------|
| $H > 1/2$ | Positive | **Persistent / trending** | Momentum markets, emerging markets |
| $H = 1/2$ | Zero | **No memory** | Efficient market (standard BM) |
| $H < 1/2$ | Negative | **Anti-persistent / mean-reverting** | FX markets, high-frequency price series |

The Hurst exponent can be estimated empirically from time series data using:
- **R/S analysis** (Hurst's original method)
- **Detrended Fluctuation Analysis (DFA)**
- **Wavelet-based estimators**

Empirical estimates of $H$ for equity index returns typically give $H \approx 0.50$–$0.56$, suggesting very mild persistence at daily frequency. Much stronger long memory ($H \approx 0.7$–$0.9$) is found in **realized volatility** series.

---

### Step 5 — Verify $H = 1/2$ recovers standard Brownian Motion

Setting $H = 1/2$:
$$\mathbb{E}[B_{1/2}(t)B_{1/2}(s)] = \frac{1}{2}\bigl(t + s - |t - s|\bigr)$$

For $t \ge s \ge 0$: $|t - s| = t - s$, so:
$$= \frac{1}{2}\bigl(t + s - (t-s)\bigr) = \frac{1}{2}(2s) = s = \min(t, s)$$

The covariance $\text{Cov}(W_t, W_s) = \min(t,s)$ is the defining property of standard Brownian Motion. $\square$

---

### Step 6 — Why fBm is not an Itô process (non-Markovian)

Standard BM satisfies the Markov property because its increments are independent. fBm with $H \ne 1/2$ has **correlated increments**, which means the conditional distribution of $B_H(t)$ depends not just on the current value but on the entire history. It is therefore **not Markovian** and **cannot be written as a standard Itô process** $dx = a(x,t)\,dt + b(x,t)\,dW$.

This has practical consequences:
- fBm models are more complex to work with (no standard stochastic calculus).
- The risk-neutral pricing framework does not directly apply.
- Hedging strategies based on delta-hedging need modification.

Rough volatility models (e.g., the **Rough Bergomi** model) use fBm with $H \approx 0.1$ for the log-volatility, capturing the empirically observed roughness of implied volatility surfaces.
