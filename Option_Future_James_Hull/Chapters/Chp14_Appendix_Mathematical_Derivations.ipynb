The derivations follow this logical chain:

## 1. The Markov Property

**Definition.** A stochastic process $\{X_t,\, t \ge 0\}$ is a **Markov process** if, for any times $t_0 < t_1 < \cdots < t_n < t$ and any value $x$:
$$\mathbb{P}\bigl(X_t \le x \mid X_{t_n}, X_{t_{n-1}}, \ldots, X_{t_0}\bigr) = \mathbb{P}\bigl(X_t \le x \mid X_{t_n}\bigr)$$

In words: the entire history of the process up to time $t_n$ is fully summarised by the current state $X_{t_n}$. Knowing older values $X_{t_{n-1}}, \ldots, X_{t_0}$ adds no further predictive power.

### Key consequence: variance is additive over disjoint intervals

Let $X$ be a Markov process. Consider the change over two non-overlapping intervals:
$$\Delta X_{[0,1]} = X_1 - X_0, \qquad \Delta X_{[1,2]} = X_2 - X_1$$

Because the process is Markov, $\Delta X_{[1,2]}$ depends only on $X_1$ and not on the path before time 1. This makes $\Delta X_{[0,1]}$ and $\Delta X_{[1,2]}$ **statistically independent**. For independent random variables:
$$\text{Var}(X_2 - X_0) = \text{Var}(\Delta X_{[0,1]} + \Delta X_{[1,2]}) = \text{Var}(\Delta X_{[0,1]}) + \text{Var}(\Delta X_{[1,2]})$$

If the variance rate per unit time is $\sigma^2$, then:
$$\text{Var}(X_t - X_0) = \sigma^2 t \implies \text{Std}(X_t - X_0) = \sigma \sqrt{t}$$

This is where the famous $\sqrt{t}$ scaling of uncertainty comes from.

## 2. Constructing the Wiener Process as the Limit of a Random Walk

### Step 1 — Set up the discrete random walk

Partition the time horizon $[0, T]$ into $N$ equal steps of length $\Delta t = T/N$. At each step $i$, a variable moves by a random amount $X_i$:
$$X_i = \begin{cases} +\Delta x & \text{with probability } \tfrac{1}{2} \\ -\Delta x & \text{with probability } \tfrac{1}{2} \end{cases}$$

**Mean of one step:**
$$\mathbb{E}[X_i] = \tfrac{1}{2}(+\Delta x) + \tfrac{1}{2}(-\Delta x) = 0$$

**Second moment of one step:**
$$\mathbb{E}[X_i^2] = \tfrac{1}{2}(\Delta x)^2 + \tfrac{1}{2}(\Delta x)^2 = (\Delta x)^2$$

**Variance of one step** (using $\text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2$):
$$\text{Var}(X_i) = (\Delta x)^2 - 0^2 = (\Delta x)^2$$

### Step 2 — Sum the steps over the total time $T$

The total accumulated change is $W_T = \sum_{i=1}^{N} X_i$, where $N = T/\Delta t$.

**Mean of the total walk:**
$$\mathbb{E}[W_T] = \sum_{i=1}^{N} \mathbb{E}[X_i] = N \cdot 0 = 0$$

**Variance of the total walk** (independence of steps → variances add):
$$\text{Var}(W_T) = \sum_{i=1}^{N} \text{Var}(X_i) = N \cdot (\Delta x)^2 = \frac{T}{\Delta t} \cdot (\Delta x)^2$$

### Step 3 — Take the continuous limit $\Delta t \to 0$

As $\Delta t \to 0$, the number of steps $N \to \infty$. For the variance to remain finite and equal to $T$ in the limit, we need:
$$\frac{T}{\Delta t} \cdot (\Delta x)^2 = T \implies \frac{(\Delta x)^2}{\Delta t} = 1 \implies (\Delta x)^2 = \Delta t$$
$$\therefore \quad \Delta x = \pm\sqrt{\Delta t}$$

With this scaling, each individual step shrinks as $\sqrt{\Delta t}$ (not linearly with $\Delta t$), which is why Brownian paths are much more volatile than deterministic paths when viewed at fine time scales.

### Step 4 — Apply the Central Limit Theorem

With the choice $\Delta x = \pm\sqrt{\Delta t}$, $W_T = \sum_{i=1}^{N} X_i$ is a sum of $N \to \infty$ independent, identically distributed random variables with mean 0 and variance $\Delta t$ each. By the **Central Limit Theorem**, this sum converges in distribution to:
$$W_T \xrightarrow{\,d\,} \mathcal{N}\bigl(0,\; N \cdot \Delta t\bigr) = \mathcal{N}(0, T)$$

### Step 5 — Write the continuous-time differential

In the limit, the increment in an infinitesimal time interval $dt$ is written as:
$$dW = \epsilon\,\sqrt{dt}, \quad \epsilon \sim \mathcal{N}(0,1)$$

This means:
$$\mathbb{E}[dW] = 0, \qquad \text{Var}(dW) = \mathbb{E}[dW^2] = dt$$

And any finite-horizon change satisfies:
$$W_T - W_0 = \int_0^T dW \sim \mathcal{N}(0, T)$$

## 3. Generalized Wiener Process

The standard Wiener process $dW$ has zero drift and a variance rate of exactly 1. To model a broader class of variables we introduce:
$$dx = a\,dt + b\,dW$$
where $a$ is the **drift rate** (units: variable per unit time) and $b$ is the **diffusion coefficient** (units: variable per $\sqrt{\text{time}}$).

### Step 1 — Integrate to find $x(T)$

Integrating both sides from $0$ to $T$:
$$\int_0^T dx = \int_0^T a\,dt + \int_0^T b\,dW$$
$$x(T) - x(0) = a\,T + b\,W_T$$

### Step 2 — Compute the expected value

Take expectations on both sides (using $\mathbb{E}[W_T] = 0$):
$$\mathbb{E}[x(T) - x(0)] = a\,T + b\,\mathbb{E}[W_T] = a\,T + b \cdot 0 = aT$$

### Step 3 — Compute the variance

Since $a\,T$ is a constant (deterministic), it contributes zero variance:
$$\text{Var}(x(T) - x(0)) = \text{Var}(b\,W_T) = b^2\,\text{Var}(W_T) = b^2 T$$

### Step 4 — State the distribution

Since $W_T \sim \mathcal{N}(0,T)$ and $x(T) - x(0) = aT + b W_T$ is an affine transformation of a normal variable:
$$x(T) - x(0) \sim \mathcal{N}(aT,\; b^2 T)$$

For a short interval $\Delta t$, the analogous discrete-time approximation is:
$$\Delta x = a\,\Delta t + b\,\epsilon\sqrt{\Delta t}, \quad \epsilon \sim \mathcal{N}(0,1)$$

## 4. Derivation of Itô's Lemma via Taylor Expansion

**Setup.** Let $x$ follow the Itô process:
$$dx = a(x,t)\,dt + b(x,t)\,dW$$

Let $G(x, t)$ be a twice continuously differentiable function. We want to find $dG$.

### Step 1 — Write the full 2D Taylor expansion

$$G(x + dx,\; t + dt) = G(x, t) + \frac{\partial G}{\partial x}dx + \frac{\partial G}{\partial t}dt
+ \frac{1}{2}\frac{\partial^2 G}{\partial x^2}(dx)^2 + \frac{1}{2}\frac{\partial^2 G}{\partial t^2}(dt)^2
+ \frac{\partial^2 G}{\partial x\,\partial t}\,dx\,dt + \cdots$$

Therefore:
$$dG = \frac{\partial G}{\partial x}dx + \frac{\partial G}{\partial t}dt + \frac{1}{2}\frac{\partial^2 G}{\partial x^2}(dx)^2
+ \underbrace{\frac{1}{2}\frac{\partial^2 G}{\partial t^2}(dt)^2 + \frac{\partial^2 G}{\partial x\,\partial t}\,dx\,dt}_{\text{higher-order}} + \cdots$$

### Step 2 — Expand $(dx)^2$

Substitute $dx = a\,dt + b\,dW$:
$$(dx)^2 = (a\,dt + b\,dW)^2 = a^2(dt)^2 + 2ab\,dt\,dW + b^2(dW)^2$$

### Step 3 — Evaluate each second-order term as $dt \to 0$

This is the heart of stochastic calculus. We need the order of magnitude of each term:

| Term | Expression | Order | Limit |
|---|---|---|---|
| $(dt)^2$ | $(dt)^2$ | $O(dt^2)$ | $\to 0$ |
| $dt\,dW$ | $dt \cdot \epsilon\sqrt{dt} = \epsilon\,dt^{3/2}$ | $O(dt^{3/2})$ | $\to 0$ |
| $(dW)^2$ | $\epsilon^2 \cdot dt$ | $O(dt)$ | **$\to dt$** |

The critical step is understanding why $(dW)^2 \to dt$ (rather than going to zero). Because $dW = \epsilon\sqrt{dt}$, we have:
$$(dW)^2 = \epsilon^2 \cdot dt$$

The variable $\epsilon^2$ has mean $\mathbb{E}[\epsilon^2] = 1$ (since $\epsilon \sim \mathcal{N}(0,1)$ and $\mathbb{E}[\epsilon^2] = \text{Var}(\epsilon) = 1$) and its own variance is:
$$\text{Var}(\epsilon^2) = \mathbb{E}[\epsilon^4] - (\mathbb{E}[\epsilon^2])^2 = 3 - 1 = 2$$

Therefore $(dW)^2 = \epsilon^2 dt$ has:
- Mean: $\mathbb{E}[(dW)^2] = dt$
- Variance of $(dW)^2$: $\text{Var}(\epsilon^2 dt) = 2(dt)^2 \to 0$ as $dt \to 0$

Since the variance of $(dW)^2$ itself goes to zero as $dt \to 0$, the random variable $(dW)^2$ **degenerates to its mean** $dt$ in the limit. This is made rigorous by the theory of **quadratic variation**:
$$\text{q.v.}(W)_t = \lim_{\|\Pi\| \to 0} \sum_i (W_{t_{i+1}} - W_{t_i})^2 = t \quad \text{(a.s.)}$$

Therefore the stochastic multiplication table is:
$$dW \cdot dW = dt, \qquad dt \cdot dW = 0, \qquad dt \cdot dt = 0$$

### Step 4 — Substitute back into $(dx)^2$

$$(dx)^2 = a^2 \underbrace{(dt)^2}_{0} + 2ab\underbrace{\,dt\,dW}_{0} + b^2\underbrace{(dW)^2}_{dt} = b^2\,dt$$

### Step 5 — Substitute $dx$ and $(dx)^2$ into the Taylor expansion

Drop all higher-order terms (they go to zero faster than $dt$):
$$dG = \frac{\partial G}{\partial x}(a\,dt + b\,dW) + \frac{\partial G}{\partial t}dt + \frac{1}{2}\frac{\partial^2 G}{\partial x^2}(b^2\,dt)$$

### Step 6 — Collect $dt$ and $dW$ terms

Expand the bracket and group:
$$dG = \frac{\partial G}{\partial x}a\,dt + \frac{\partial G}{\partial x}b\,dW + \frac{\partial G}{\partial t}dt + \frac{1}{2}\frac{\partial^2 G}{\partial x^2}b^2\,dt$$

$$\boxed{dG = \left(\frac{\partial G}{\partial x}a + \frac{\partial G}{\partial t} + \frac{1}{2}\frac{\partial^2 G}{\partial x^2}b^2\right)dt + \frac{\partial G}{\partial x}b\,dW}$$

This is **Itô's Lemma**. The term $\frac{1}{2}\frac{\partial^2 G}{\partial x^2}b^2$ is entirely absent from ordinary calculus and arises solely because $(dW)^2 = dt \ne 0$.

## 5. Exact Solution of Geometric Brownian Motion

The standard model for a non-dividend-paying stock price is:
$$dS = \mu S\,dt + \sigma S\,dW \quad \text{(Geometric Brownian Motion)}$$

We solve this SDE explicitly using Itô's Lemma.

### Step 1 — Choose the transformation $G(S,t) = \ln S$

In ordinary calculus, $d(\ln S) = dS/S$. In stochastic calculus this is not quite right. We apply Itô's Lemma with $x = S$, $a = \mu S$, $b = \sigma S$:

Compute the required partial derivatives:
$$\frac{\partial G}{\partial S} = \frac{\partial}{\partial S}\ln S = \frac{1}{S}$$
$$\frac{\partial^2 G}{\partial S^2} = \frac{\partial}{\partial S}\left(\frac{1}{S}\right) = -\frac{1}{S^2}$$
$$\frac{\partial G}{\partial t} = 0 \quad (\ln S \text{ does not depend explicitly on } t)$$

### Step 2 — Plug into Itô's Lemma

$$d(\ln S) = \left(\frac{1}{S}\cdot\mu S + 0 + \frac{1}{2}\cdot\left(-\frac{1}{S^2}\right)\cdot\sigma^2 S^2\right)dt + \frac{1}{S}\cdot\sigma S\,dW$$

Simplify each term:
$$d(\ln S) = \left(\mu - \frac{\sigma^2}{2}\right)dt + \sigma\,dW$$

### Step 3 — Interpret the result

Notice that the right-hand side has **constant coefficients** (no dependence on $S$ or $t$). This is exactly a Generalized Wiener Process for $G = \ln S$, and we can integrate it directly.

### Step 4 — Integrate both sides from $0$ to $T$

$$\int_0^T d(\ln S) = \int_0^T \left(\mu - \frac{\sigma^2}{2}\right)du + \int_0^T \sigma\,dW_u$$
$$\ln S_T - \ln S_0 = \left(\mu - \frac{\sigma^2}{2}\right)T + \sigma W_T$$
$$\ln\frac{S_T}{S_0} = \left(\mu - \frac{\sigma^2}{2}\right)T + \sigma W_T$$

### Step 5 — Exponentiate to recover $S_T$

$$S_T = S_0 \exp\!\left[\left(\mu - \frac{\sigma^2}{2}\right)T + \sigma W_T\right]$$

Since $W_T \sim \mathcal{N}(0, T)$, the log-return is:
$$\ln\frac{S_T}{S_0} \sim \mathcal{N}\!\left(\left(\mu - \frac{\sigma^2}{2}\right)T,\; \sigma^2 T\right)$$

and $S_T$ itself follows a **lognormal distribution**.

### Step 6 — The variance drag $-\sigma^2/2$

Why is the mean of $\ln(S_T/S_0)$ equal to $(\mu - \sigma^2/2)T$ rather than $\mu T$? This is Jensen's Inequality at work. Because the exponential function is convex:
$$\mathbb{E}[S_T] = S_0 e^{\mu T} \quad \text{but} \quad \mathbb{E}[\ln S_T] = \ln S_0 + \left(\mu - \frac{\sigma^2}{2}\right)T$$

The term $-\sigma^2/2$ is the cost of being in a stochastic world — it is the "drag" imposed by the randomness itself. Higher volatility lowers the geometric mean return, even if it does not lower the arithmetic expected price.

## 6. Fractional Brownian Motion: Covariance Structure and Hurst Index

### Definition

Fractional Brownian Motion $B_H(t)$ is a zero-mean Gaussian process such that:
1. $B_H(0) = 0$ (starts at the origin)
2. It has **stationary increments**: the distribution of $B_H(t+s) - B_H(t)$ depends only on $s$, not on $t$.
3. It is **self-similar** with Hurst index $H \in (0,1)$: for any $c > 0$,
   $\{B_H(ct),\, t \ge 0\} \overset{d}{=} \{c^H B_H(t),\, t \ge 0\}$

### Step 1 — Use self-similarity to determine the variance of increments

By property 3 with $c = t$ and $s=1$:
$$B_H(t) \overset{d}{=} t^H B_H(1)$$
Therefore:
$$\text{Var}(B_H(t)) = t^{2H}\,\text{Var}(B_H(1)) = t^{2H}$$
(normalising so that $\text{Var}(B_H(1)) = 1$).

By stationarity of increments (property 2):
$$\text{Var}(B_H(t) - B_H(s)) = |t - s|^{2H}$$

### Step 2 — Expand the variance of the increment

For any two random variables $A$ and $B$:
$$\text{Var}(A - B) = \mathbb{E}[(A-B)^2] - (\mathbb{E}[A-B])^2$$

Since $\mathbb{E}[B_H(t)] = 0$ for all $t$:
$$\text{Var}(B_H(t) - B_H(s)) = \mathbb{E}[(B_H(t) - B_H(s))^2]$$

Expand the square:
$$\mathbb{E}[(B_H(t) - B_H(s))^2] = \mathbb{E}[B_H(t)^2] - 2\,\mathbb{E}[B_H(t)B_H(s)] + \mathbb{E}[B_H(s)^2]$$

We already know $\mathbb{E}[B_H(t)^2] = \text{Var}(B_H(t)) = t^{2H}$ and similarly $\mathbb{E}[B_H(s)^2] = s^{2H}$.

Substituting:
$$|t-s|^{2H} = t^{2H} - 2\,\mathbb{E}[B_H(t)B_H(s)] + s^{2H}$$

### Step 3 — Rearrange to find the covariance

Isolate $\mathbb{E}[B_H(t)B_H(s)]$:
$$2\,\mathbb{E}[B_H(t)B_H(s)] = t^{2H} + s^{2H} - |t-s|^{2H}$$

$$\boxed{\mathbb{E}[B_H(t)B_H(s)] = \frac{1}{2}\left(|t|^{2H} + |s|^{2H} - |t-s|^{2H}\right)}$$

### Step 4 — The correlation between increments

Consider two consecutive, non-overlapping unit increments $\Delta_1 = B_H(1) - B_H(0)$ and $\Delta_2 = B_H(2) - B_H(1)$.

Their covariance is:
$$\text{Cov}(\Delta_1, \Delta_2) = \mathbb{E}[\Delta_1 \Delta_2]$$

Expand:
$$= \mathbb{E}[(B_H(1) - B_H(0))(B_H(2) - B_H(1))]$$
$$= \mathbb{E}[B_H(1)B_H(2)] - \mathbb{E}[B_H(1)^2] - \mathbb{E}[B_H(0)B_H(2)] + \mathbb{E}[B_H(0)B_H(1)]$$

Using the covariance formula with $(t,s) = (1,2), (1,1), (0,2), (0,1)$:
$$\mathbb{E}[B_H(1)B_H(2)] = \frac{1}{2}(1 + 2^{2H} - 1) = \frac{2^{2H}}{2}$$
$$\mathbb{E}[B_H(1)^2] = 1$$
$$\mathbb{E}[B_H(0)B_H(2)] = 0 \quad (B_H(0) = 0)$$
$$\mathbb{E}[B_H(0)B_H(1)] = 0 \quad (B_H(0) = 0)$$

Therefore:
$$\text{Cov}(\Delta_1, \Delta_2) = \frac{2^{2H}}{2} - 1 = \frac{4^H - 2}{2}$$

The **sign** of this covariance depends entirely on $H$:
- $H > 0.5$: $4^H > 2 \implies$ covariance is **positive** (increments are positively correlated, momentum)
- $H = 0.5$: $4^{0.5} = 2 \implies$ covariance is **zero** (standard Brownian Motion, independent increments)
- $H < 0.5$: $4^H < 2 \implies$ covariance is **negative** (increments are negatively correlated, mean-reversion)

### Step 5 — Verify that $H = 0.5$ recovers standard Brownian Motion

Substituting $H = 0.5$ into the covariance formula ($2H = 1$):
$$\mathbb{E}[B_{0.5}(t)B_{0.5}(s)] = \frac{1}{2}\bigl(|t| + |s| - |t-s|\bigr)$$

For $t > s > 0$:
$$= \frac{1}{2}\bigl(t + s - (t-s)\bigr) = \frac{1}{2}(2s) = s = \min(t,s)$$

The covariance $\text{Cov}(W_t, W_s) = \min(t,s)$ is exactly the defining property of the standard Wiener process, confirming the generalisation is consistent.
